# DimCLi: retrieving documentation with the `dsl_docs` magic 
The `dsl_docs` magic prints out information about fields available for the various Dimensions API sources and entities. 

This information is retrieved dynamically using the [describe](https://docs.dimensions.ai/dsl/functions.html) DSL command, then displayed in the notebook using pandas. 

```python
import dimcli
%dsl_login
```

If you pass a wrong object name, the full list of available sources and entities is printed.

```python
%dsl_docs blabla
```

Eg. passing a source name like `grants` shows all fields available for that source. 

```python
%dsl_docs grants
```

Same for entities eg `countries`

```python
%dsl_docs countries
```

If no object is requested explicitally, the documentation for all the sources gets returned.

```python
%dsl_docs 
```
